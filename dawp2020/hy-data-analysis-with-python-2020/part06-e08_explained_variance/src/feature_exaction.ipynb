```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
X, y = load_diabetes(True)
print(X.shape)
print(y.shape)
```

```python
pca=PCA(2)
pca.fit(X)
print(pca.explained_variance_)
print(pca.components_)
Z=pca.transform(X)
plt.axis('equal')
plt.scatter(Z[:,0], Z[:,1])
```

```python
rng = np.random.RandomState(0)
X = rng.randn(3, 400)
p = rng.rand(10,3)  # Random projection into 10d
X = np.dot(p, X)
print(X)
```

```python
pca=PCA()
pca.fit(X)
v=pca.explained_variance_
print(v)
plt.plot(np.arange(1,11), v) 
plt.plot(np.arange(1,11), np.cumsum(v)) # cumulative sum of array
```
